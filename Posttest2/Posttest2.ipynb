### 4.  Tampilkan jumlah label dari semua attribute yang bernilai object (contoh : Jenis Kelamin, Agama).

```python

# Memilih atribut yang bertipe object (biasanya tipe string atau kategori)
atribut_object = wp.select_dtypes(include=['object'])

# Menghitung jumlah nilai non-null dari setiap atribut bertipe object
jumlah_nilai_non_null_per_atribut = atribut_object.count()

# Tampilkan hasil
print("Jumlah nilai non-null dari setiap atribut bertipe object:")
print(jumlah_nilai_non_null_per_atribut)
```

### 5. Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

```python
# Mengelompokkan data berdasarkan benua dan menjumlahkan populasi untuk setiap benua
continent_population = wp.groupby('Continent')['2022 Population'].sum()

# Membuat bar chart
plt.figure(figsize=(10, 6))
continent_population.plot(kind='bar', color='skyblue', edgecolor='black')

# Menambahkan judul dan label
plt.title('Total Populasi per Benua (2022)', fontsize=16)
plt.xlabel('Benua', fontsize=14)
plt.ylabel('Populasi (Miliar)', fontsize=14)

# Menampilkan grafik
plt.xticks(rotation=45)
plt.show()
```

### 6. Tampilkan korelasi dataset menggunakan heatmap.

```python
# Menghitung matriks korelasi dari dataset
korelasi = wp.corr()

# Membuat heatmap untuk matriks korelasi dengan matplotlib
plt.figure(figsize=(12, 10))
cax = plt.matshow(korelasi, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(cax)

# Menambahkan label untuk sumbu
plt.xticks(ticks=np.arange(len(korelasi.columns)), labels=korelasi.columns, rotation=90, fontsize=8)
plt.yticks(ticks=np.arange(len(korelasi.columns)), labels=korelasi.columns, fontsize=8)

# Menambahkan nilai korelasi di atas heatmap dengan ukuran font kecil
for (i, j), val in np.ndenumerate(korelasi):
    plt.text(j, i, f'{val:.2f}', ha='center', va='center', color='black', fontsize=6)

plt.title('Heatmap Korelasi Dataset', fontsize=10)
plt.show()
```
